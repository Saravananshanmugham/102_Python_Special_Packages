```python
import pymysql.cursors
import pandas as pd
```

```python
# Connect to the database
dbcon = pymysql.connect(host='localhost',
                             user='root',
                             password='K@090403$s',
                             database='pytables',
                             cursorclass=pymysql.cursors.DictCursor)
```

```python
# Cursor to the database
Cursor = dbcon.cursor()
print("Database Cursor Created Successfully")
```

```python
myquery = (
"select a.CustomerName, b.InvoiceNumber, c.Quantity, c.UnitCost, c.UnitPrice "
"from CustomerMaster a, InvoiceHeader b, InvoiceDetails c "
"where "
"a.CustomerID=b.CustomerID and "
"b.invoiceNumber =c.InvoiceNumber"
)
```

```python
# Use this cell when connection cursor is created using class pymysql.cursors.DictCursor
# fetchall_unbuffered - returns a iterator not data

Cursor.execute(myquery)
df = pd.DataFrame(Cursor.fetchall())
df.columns = ['CustomerID', 'InvoiceNumber','Quantity','UnitCost','UnitPrice']
```

```python
df
```

```python
dbcon.close()
```

### pymysql.cursors.SSCursor -- Unbuffered Cursor

```python
# pymysql.cursors.SSCursor -- Unbuffered Cursor
dbcon = pymysql.connect(host='localhost',
                             user='root',
                             password='K@090403$s',
                             database='pytables',
                             cursorclass=pymysql.cursors.SSCursor)
```

```python
# Cursor to the database
Cursor = dbcon.cursor()
print("Database Cursor Created Successfully")
```

```python
myquery = (
"select a.CustomerName, b.InvoiceNumber, c.Quantity, c.UnitCost, c.UnitPrice "
"from CustomerMaster a, InvoiceHeader b, InvoiceDetails c "
"where "
"a.CustomerID=b.CustomerID and "
"b.invoiceNumber =c.InvoiceNumber"
)
```

```python
# Use this cell when connection cursor is created using class pymysql.cursors.SSCursor
# fetchall_unbuffered - returns a iterator not data

Cursor.execute(myquery)
myresults= Cursor.fetchall_unbuffered()
```

```python
type(myresults)
```

```python
for i in myresults:
    print(i)
```

```python
dbcon.close()
```
